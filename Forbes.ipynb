## Distribution of Net Worth Among Billionaires:

We will line plot of the `net_worth` by `rank` to see the distribution of the net worth of the billionares.

```python
from matplotlib.ticker import MultipleLocator

#line plot of net_worth

plt.figure(figsize=(35, 10))
plt.plot(df['net_worth'])
plt.xlabel('rank')
plt.ylabel('net_worth')
plt.title('Net Worth of Billionaires')

plt.gca().xaxis.set_major_locator(MultipleLocator(50))
plt.gca().yaxis.set_major_locator(MultipleLocator(10))

plt.show()
```

It was quite surprizing to see that this graph is not linear. It is exponential. It means that the net worth of the billionares is not distributed linearly. At the rank 50 we can see `knee point` of the graph.

There are just `50` people among `2640` who have net worth greater that approx `27 billions`.

### Bar graphs of Freaquency, Cumulative and Average `net_worth` of billionaires by `Country` and `Industry`

```python
country_counts = df['country'].value_counts() #Datatype: Series, Index: country, values: frequency

country_counts = country_counts[:30]
country_counts = country_counts[::-1]
plt.figure(figsize=(8, 6))

plt.barh(country_counts.index, country_counts.values)

plt.xlabel('Frequency')
plt.ylabel('countries')
plt.title('Frequency of countries')
plt.show()
```

```python
country_net_worth = df.groupby('country')['net_worth'].sum()
country_net_worth = country_net_worth.sort_values(ascending=False)
country_net_worth = country_net_worth[:30]
country_net_worth = country_net_worth[::-1]

plt.figure(figsize=(8, 6))

plt.barh(country_net_worth.index, country_net_worth.values)

plt.xlabel('Net Worth')
plt.ylabel('countries')
plt.title(' Cumulative Net Worth of countries')
plt.show()
```

```python
country_avg_net_worth = df.groupby('country')['net_worth'].mean() #Datatype: Series, Index: country, values: mean of net_worth
country_avg_net_worth = country_avg_net_worth.sort_values(ascending=False)
country_avg_net_worth = country_avg_net_worth[:30]
country_avg_net_worth = country_avg_net_worth[::-1]

plt.figure(figsize=(8, 6))

plt.barh(country_avg_net_worth.index, country_avg_net_worth.values)

plt.xlabel('Average Net Worth')
plt.ylabel('countries')
plt.title('Average Net Worth of countries')
plt.show()
```

```python
country_counts = df['industry'].value_counts().sort_values(ascending=True)

plt.barh(country_counts.index, country_counts.values)
plt.xlabel('Frequency')
plt.ylabel('industries')
plt.title('Frequency of industy')
plt.show()
```

```python
industry_net_worth = df.groupby('industry')['net_worth'].sum()
industry_net_worth = industry_net_worth.sort_values(ascending=True)

plt.barh(industry_net_worth.index, industry_net_worth.values)
plt.xlabel('Net Worth')
plt.ylabel('industries')
plt.title('Cumulative Net Worth of industy')
plt.show()
```

```python
industry_avg_net_worth = df.groupby('industry')['net_worth'].mean()
industry_avg_net_worth = industry_avg_net_worth.sort_values(ascending=True)

plt.barh(industry_avg_net_worth.index, industry_avg_net_worth.values)
plt.xlabel('Average Net Worth')
plt.ylabel('industries')
plt.title('Average Net Worth of industy')
plt.show()
```

#### finding number of outliers in both Inherited and Self-Made wealth

##### Self-Made:

```python
q1 = df[df['Self Made'] == 1]['net_worth'].quantile(0.25)
q3 = df[df['Self Made'] == 1]['net_worth'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f'Number of outliers: {len(df[(df["Self Made"] == 1) & ((df["net_worth"] < lower_bound) | (df["net_worth"] > upper_bound))])}')
```

##### Inherited:

```python
q1 = df[df['Self Made'] == 0]['net_worth'].quantile(0.25)
q3 = df[df['Self Made'] == 0]['net_worth'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f'Number of outliers: {len(df[(df["Self Made"] == 0) & ((df["net_worth"] < lower_bound) | (df["net_worth"] > upper_bound))])}')
```

## Education Analysis:

```python
import matplotlib.pyplot as plt
import seaborn as sns

education_df = df[['Self Made', 'Bachelor', 'Master', 'Doctorate', 'Drop Out']]

education_counts = education_df.groupby('Self Made').sum()
print(education_counts, '\n\n')

# Transposing `education_counts` for better visualization
education_counts = education_counts.transpose()
print(education_counts, '\n\n')

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', stacked=False, cmap='viridis')
education_counts.plot(kind='line', stacked=False, cmap='viridis')
plt.title('Education Differences between Self-Made and Inherited Billionaires')
plt.xlabel('Education Level')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=0)
plt.legend(title='Self Made', loc='upper right', labels=['Inherited', 'Self-Made'])
plt.tight_layout()
plt.show()
```

```python
# Count the number of billionaires with each education level
education_counts = df[['Bachelor', 'Master', 'Doctorate', 'Drop Out']].sum()

# Visualize the distribution of educational levels
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', cmap='viridis')
plt.title('Educational Background of Billionaires')
plt.xlabel('Education Level')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)
plt.show()
```
